```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
from PIL import Image
```

```python
# Define your data directories
train_dir = 'nu/train/'
validation_dir = 'nu/test/'
```

```python
# Convert images and CSV data into numpy arrays

print('--------Training--------')

train_images = []
train_labels = []
for filename in os.listdir(train_dir+'images/'):
    if len(filename) == len('Fuente0_parte_amistoso_0.jpg'):
        continue
    print(filename)
    img = Image.open(train_dir+'images/'+filename)
    train_images.append(np.array(img))
    train_labels.append(pd.read_csv(train_dir+'labels/'+filename[:-4]+'.csv').to_numpy())
print("Generating numpy arrays...")
train_images = np.array(train_images)
train_labels = np.array(train_labels)

print('--------Validation--------')

validation_images = []
validation_labels = []
for filename2 in os.listdir(validation_dir+'images/'):
    if len(filename2) == len('Fuente0_parte_amistoso_0.jpg') or len(filename2) == len('Fuente10_parte_amistoso_0.jpg'):
        continue
    print(filename2)
    img2 = Image.open(validation_dir+'images/'+filename2)
    validation_images.append(np.array(img2))
    validation_labels.append(pd.read_csv(validation_dir+'labels/'+filename2[:-4]+'.csv').to_numpy())
print("Generating numpy arrays...")
validation_images = np.array(validation_images)
validation_labels = np.array(validation_labels)
print("Done!")
```

```python
# Define your model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10)
])
```

```python
# Compile your model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
# Train your model
model.fit(train_images.astype('float32'), train_labels, batch_size=16, epochs=1)
```

```python
# Evaluate your model
test_loss, test_acc = model.evaluate(validation_images, validation_labels, verbose=2)
print('\nTest accuracy:', test_acc)
```
