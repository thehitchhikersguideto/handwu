```python
# Imports
import pandas as pd
import numpy as np
import os
import glob
import warnings
import csv
import shutil

warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Instantiate a dataframe 
df = pd.DataFrame(columns=['folder', 'file_name', 'text'])
```

```python
directory_path = '../session_12_Group_Work/'

for folder in os.listdir(directory_path):
    if folder.startswith('.'):
        continue
    for file in os.listdir(directory_path + folder):
        if file.endswith('.csv') or file.startswith('.') or len(file) == len('Fuente0_parte_amistoso_0.jpg') or len(file) == len('Fuente10_parte_amistoso_0.jpg'):
            continue
        else:
            df = df.append({'folder': folder, 'file_name': file}, ignore_index=True)
```

```python
csv_dir = '../cleaned_csv/'
os.makedirs(csv_dir, exist_ok=True)

# Clean csv files
for folder in os.listdir(directory_path):
    if folder.startswith('.'):
        continue
    for file in os.listdir(directory_path + folder):
        if file.endswith('.csv'):
            if file.startswith('.'):
                continue
            else:
                shutil.copy(directory_path + folder + '/' + file, csv_dir + file)
```

```python
for file in os.listdir(csv_dir):
    if file.endswith('.csv'):
        df2 = pd.read_csv(os.path.join(csv_dir, file))
        df2.drop(df2.columns[[0, 1]], axis=1, inplace=True)
        df2.to_csv(os.path.join(csv_dir, file), index=False)
```

```python
def get_text_from_csv(file): 
    with open(file, 'r') as f:
        reader = csv.reader(f)
        # convert reader obejct to body of text
        text = ''
        for row in reader:
            text += ' '.join(row)
        return text

def change_file_name_string(file):
    file = file.replace('.jpg', '.csv')
    return file


def add_text_from_csv_to_df(df):
    for index, row in df.iterrows():
        file = change_file_name_string(row['file_name'])
        
        PATH = f'{csv_dir}{file}'
        # if path exists 
        if os.path.exists(PATH):
            text = get_text_from_csv(PATH)
            df.loc[index, 'text'] = text
            # df.loc[index, 'text_original'] = file used this to check that the csv and the jpgs matched
    print(f'Showing the dataframe after adding the text from the csv files:\n{df.head()}')
            
add_text_from_csv_to_df(df)
```

```python
def save_csv_file(df):
    df.to_csv(f'df.csv', index=False)
    print('File saved')

save_csv_file(df)
```
